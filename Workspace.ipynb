```python
ax = plt.gca() #get current axis 
fig = plt.gcf()
fig.set_size_inches(15.5,10.5)


df.plot(kind='line',x='year',y='avg_global_temp',ax=ax,color='blue')
df2.plot(kind='line',x='year',y='avg_local_temp',ax=ax,color='red')
plt.ylabel('Moving Average Temperatures (Celsius)',fontweight="bold",size='18')
plt.xlabel('Year',fontweight="bold",size='18')
plt.title('Line Chart of Global vs. Raleigh, NC  Moving Average Temperature Change over Years',fontweight="bold",size='18')
plt.legend(['Global','Raleigh, NC'],loc='center left')
plt.grid()
plt.table()
plt.savefig('temp.jpg', bbox_inches='tight') #removes whitespace around the image. )
plt.show()


```
